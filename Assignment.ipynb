```python
#Assignment 1: Plotting Charts using Matplotlib
''' Create a plot comparing Kohli, Rohit Sharma, and Sehwag across 10 years of hypothetical runs.
Use:Labels, Legends, Colors, Line styles'''

import numpy as np
import matplotlib.pyplot as plt

#years = [2000,2001,2002,2003,2004,2005,2006,2007,2008,2009]
#years = [year for year in range(2000, 2010)]
years = np.arange(2000, 2010)

Sehwag = [200, 400, 600, 800, 1000, 1200, 1400, 1600, 1400, 1200]
kohli = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1400]
Rohit_Sharma = np.random.randint(0, 2001, size=10)
Dhoni = np.random.randint(0, 2001, size=10) 

plt.plot(years, kohli,color="red", linestyle='-',label="Kholi")
plt.plot(years, Rohit_Sharma, color="blue",linestyle="--", label="Rohit Sharma")
plt.plot(years, Sehwag, color="green", linestyle="-.",label="Sehwag")
plt.plot(years, Dhoni, color="black", linestyle=":", label="Dhoni")

plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Run Comparison")

plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
#Assignment 2: Bar Chart in Matplotlib
'''Create a side-by-side bar chart comparing runs of Sachin, Kohli, and Sehwag over 5 selected years.
Then create a horizontal bar chart showing total runs scored in their debut 5 years.'''

years = np.arange(1991, 1996)
Sachin = np.random.randint(0, 1501, size=5)
Kohli = np.random.randint(0, 1501, size=5)
Sehwag = np.random.randint(0, 2001, size=5)

x = np.arange(len(years))
width = 0.25

plt.bar(x-width, Sachin, color='red', width=width, label='Sachin Runs')
plt.bar(x, Kohli, color='skyblue', width=width, label='Kohli Runs')
plt.bar(x+width ,Sehwag, color='lightgreen', width=width, label='Sehwag Runs')

plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Runs in BarChart")

plt.xticks(x,years)
plt.legend()
plt.tight_layout()
plt.show()

#TASK 2

Players = ["Sachin", "Kohli", "Sehwag"]
Runs_5years= [100+200+300+400+500, 500+300+200+300+100, 700+100+600+100+500]
plt.barh(Players, Runs_5years, color="skyblue")
plt.xlabel("Players")
plt.ylabel("Runs in 5 Years")
plt.title("Comparison")

plt.tight_layout()
plt.show()
```

```python
#Assignment3 : Pie Chart in Matplotlib
'''Create a pie chart showing the market share of mobile OS (Android, iOS, others).'''

import matplotlib.pyplot as plt
 
plt.style.use("ggplot")

# Data
Shares = [45, 40, 10, 5]
Company = ['Android', 'iOS', 'Linux', 'others']
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
explode = [0, 0, 0.2, 0] 

plt.pie(Shares,
        labels=Company,
        colors=colors,
        explode=explode,
        shadow=True,
        autopct='%1.0f%%',
        startangle=140,
        wedgeprops={'edgecolor': 'black', 'linewidth': 2} )

plt.title("Market Share of Mobile OS")
plt.show()
```

```python
# HOW TO HANDLE CROWDED PIE CHART :BEST TO AVOID THOUGH
# Recreate the same using a horizontal bar chart and observe which is easier to understand.'''

import matplotlib.pyplot as plt

# Original data
languages = ['Python', 'Java', 'C++', 'JavaScript', 'C#', 'Ruby', 'Go', 'Rust', 'Swift', 'PHP']
usage = [30, 20, 10, 10, 7, 5, 4, 3, 2, 1]

# Set threshold for combining into "Others"
threshold = 5

# Separate major and minor languages
major_languages = []
major_usage = []
others_total = 0

for lang, use in zip(languages, usage):
    if use >= threshold:
        major_languages.append(lang)
        major_usage.append(use)
    else:
        others_total += use

# Add "Others" category if there are small values
if others_total > 0:
    major_languages.append('Others')
    major_usage.append(others_total)

# Plot
plt.pie(major_usage, labels=major_languages, autopct='%1.1f%%', startangle=90)
plt.title("Programming Language Usage")
plt.show()

print(f"Others includes: {[lang for lang, use in zip(languages, usage) if use < threshold]}")
print(f"Others total: {others_total}%")


# Horizontal bar chart
import matplotlib.pyplot as plt

# Original data
languages = ['Python', 'Java', 'C++', 'JavaScript', 'C#', 'Ruby', 'Go', 'Rust', 'Swift', 'PHP']
usage = [30, 20, 10, 10, 7, 5, 4, 3, 2, 1]

# Set threshold for combining into "Others"
threshold = 5

# Separate major and minor languages
major_languages = []
major_usage = []
others_total = 0
others_list = []

for lang, use in zip(languages, usage):
    if use >= threshold:
        major_languages.append(lang)
        major_usage.append(use)
    else:
        others_total += use
        others_list.append(lang)

# Add "Others" category if there are small values
if others_total > 0:
    major_languages.append('Others')
    major_usage.append(others_total)


#----------------------------------------- Create horizontal bar chart-------------------------------------------------------------------------------
plt.figure(figsize=(10, 6))
plt.barh(major_languages, major_usage, color='skyblue', edgecolor='black')
plt.xlabel('Usage (%)')
plt.title('Programming Language Usage')
plt.grid(axis='x', alpha=0.3)

# Add percentage labels on bars
for i, (lang, use) in enumerate(zip(major_languages, major_usage)):
    plt.text(use + 0.5, i, f'{use}%', va='center')

plt.tight_layout()
plt.show()

print(f"Others includes: {others_list}")
print(f"Others total: {others_total}%")
```

```python
#Assignment4 : Stack Plots in Matplotlib
'''Try making a stackplot with your own weekly schedule!'''

import matplotlib.pyplot as plt

activities = ['DSA' , 'Data_Science', 'Leetcode', 'Sleep', 'Entertainment', 'Walk']
time_spent=[3, 3, 2, 8, 4, 1]
colors = ['skyblue', 'lightgreen', 'gold', 'purple', 'red', 'black']

plt.figure(figsize=(8,8))
plt.pie(time_spent, labels =activities, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Time Spent')
plt.xlabel('Hours')
plt.ylabel('Tasks')
plt.legend(loc='upper left')

#-----------------------------------------------Stack Plot----------------------------------------------

import matplotlib.pyplot as plt

Days = [1, 2, 3, 4, 5, 6, 7]
DSA = [3, 4, 3, 5, 4, 3, 4]
Data_Science = [2, 2, 1, 1, 2, 3, 2]
Leetcode = [2, 1, 2, 2, 1, 1, 1]
Sleep = [5, 5, 6, 5, 6, 5, 5]
Entertainment=[3,4,5,4,5,5,3]
Walk=[2,1,1,2,2,1,1]

Activity = ['DSA' , 'Data_Science', 'Leetcode', 'Sleep', 'Entertainment', 'Walk']
colors = ['skyblue', 'lightgreen', 'gold', 'purple', 'red', 'black']


plt.figure(figsize=(10,6))
plt.stackplot(Days,DSA, Data_Science, Leetcode, Sleep, Entertainment,Walk,
              labels=Activity, colors=colors, alpha=0.8)

plt.legend(loc='upper left') 
plt.title('Weekly Activity Tracker')
plt.xlabel('Day')
plt.ylabel('Hours')
plt.grid(True)
plt.show()
```
